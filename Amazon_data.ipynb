```python
import pandas as pd
import datetime
import pandas_datareader
import pandas_datareader.data as web
from pandas import Series, DataFrame


start = datetime.datetime(2015, 9, 1)
end = datetime.datetime(2020, 8, 31)

df = web.DataReader("AMZN", 'yahoo', start, end)
df.tail()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

close_px = df['Adj Close']
moving_avg = close_px.rolling(window=100).mean() # Smoothing by taking average over specific period, here a window of hundred readings.

import matplotlib as mpl
mpl.rc('figure', figsize=(10, 10))
mpl.__version__

style.use('classic')

close_px.plot(label='AMZN')
moving_avg.plot(label='moving_avg')
plt.legend()
```
